#### 4.2 Can you separate the joined columns into year, month, and day columns?

```python
fifa_df['Joined']
```

```python
#making the dates into the right data type
#converting the months to corresponding number
def monthToNumber(x):
  if x == 'Jan':
    return 1
  elif x== 'Feb':
    return 2
  elif x== 'Mar':
    return 3
  elif x== 'Apr':
    return 4
  elif x== 'May':
    return 5
  elif x== 'Jun':
    return 6
  elif x== 'Jul':
    return 7
  elif x== 'Aug':
    return 8
  elif x== 'Sep':
    return 9
  elif x== 'Oct':
    return 10
  elif x== 'Nov':
    return 11
  elif x== 'Dec':
    return 12
  else:
    return "Error"
```

```python
# creating list called date_list to store the converted dates
date_list = []
# solitting the date and modifying the day, month, and year
for x in range(len(fifa_df["Joined"])):
  date = fifa_df["Joined"][x]
  split_date = date.split(" ")
  month = str(monthToNumber(split_date[0]))
  day = str(split_date[1].replace(',',''))
  year = str(split_date[2])

  #adding 0 to single digit dates
  if len(day) == 1:
    day = ('0'+str(day))
  #recombiing the split date and then converting to correct format
  full_date = (str(month)+'/'+str(day)+'/'+str(year))
  converted_date = pd.to_datetime(full_date)
  date_list.append(converted_date)
```

```python
fifa_df["Joined"] = date_list
fifa_df["Joined"]
```

```python
type(fifa_df["Joined"])
```

```python
fifa_df['Day'] = fifa_df["Joined"].dt.day
fifa_df['Month'] = fifa_df["Joined"].dt.month
fifa_df['Year'] = fifa_df["Joined"].dt.year
fifa_df.head()
```

As seen above,columns have the some values in thousands (K) and millions (M) Which have to converted and the currency symbols removed.

```python
def object_to_float(x):
  if isinstance(x, str):
    #If K is present at the last index
    if x.find('K') != -1:
      #replace K and € with empty string
      x = x.replace('K', '').replace('€','')
      #multiply with 1000 to ensure it's in thousands
      return float(x)*1000
     #If M is present
    elif x.find('M') != -1:
      x= x.replace('M','').replace('€','')
       #multiply with 10e6 to ensure it's in millions
      return float(x)*10e6
    elif x.find('€') != -1:
      x = x.replace('€','')
      return float(x)
    else:
      return x
  else:
    return x
```

```python
fifa_df['Value(€)'] = fifa_df['Value'].apply(object_to_float)
fifa_df['Value(€)'].unique
```

```python
fifa_df['Wage(€)'] = fifa_df['Wage'].apply(object_to_float)
fifa_df['Wage(€)'].unique
```

```python
fifa_df['Release Clause(€)'] = fifa_df['Release Clause'].apply(object_to_float)
fifa_df['Release Clause(€)'].unique
```

#### 4.6 How can you make the contract column better?

```python
fifa_df["Contract"].unique()
```

It can be observed that some players have an active contract with their clubs while others are either on loan or are free. These players' contracts can be categorized into active, on_loan and free respectively.

```python
# function to categorise contracts
def categoriseContracts(x):
  #active contracts
  if "~" in x:
    x = "active"
    return x
  #on loan
  elif "On Loan" in x:
    x = "on_loan"
    return x
  #free
  elif "Free" in x:
    x = "free"
    return x
```

```python
fifa_df['Contract'] = fifa_df['Contract'].apply(categoriseContracts)
#rename contract to contract status
fifa_df = fifa_df.rename(columns={"Contract": "Contract_Status"})
#converting to category type
fifa_df["Contract_Status"] =  fifa_df["Contract_Status"].astype("category")
fifa_df.head()
```

#### 4.7 Do the height and weight columns have the appropriate data types?

```python
fifa_df['Height'].dtypes
```

```python
fifa_df['Weight'].dtypes
```

The height and weight columns both have object data types. The Object data type is usually used to store strings and mixed values. This is beacuse the values are save using their units i.e height is saved in cm while weight in kg. These columns don't have the appropriate data types as height and weight are numerical parameters usually represented using float64 or int64 data types.
The columns need to be cleaned.

```python
fifa_df[['Height','Weight']]
```

```python
fifa_df['Height'].unique()
```

```python
fifa_df['Weight'].unique()
```

```python
# Function to convert height to cm
def heightConversion(x):
  if x[-1] == "m":
    return int(x[:-2])
  elif x[-1] == '"':
    #replacing " with empty space
    x = x.replace("\"","")
    #converting the inches part of height to cm (1 inch= 2.54cm)
    height_inches = int(x[2:])* 2.54
    #converting foot part of height to cm (1 foot= 30.48cm)
    height_foot = int(x[0])*30.48
    return round(height_foot+height_inches)
```

```python
fifa_df["Height(cm)"] = fifa_df["Height"].apply(heightConversion)
fifa_df["Height(cm)"].unique()
```

```python
fifa_df.drop("Height",axis=1, inplace=True)
```

```python

#Function to convert the weight to kg
def weightConversion(x):
  if "lbs" in x:
    x = x.replace('lbs','')
    x = float(x)*0.453592
    x = int(x)
    return x
  else:
    return int(x[:-2])
```

```python
fifa_df["Weight(kg)"] = fifa_df["Weight"].apply(weightConversion)
fifa_df["Weight(kg)"].unique()
```

```python
fifa_df.drop("Weight",axis=1, inplace=True)
```

#### 4.7 Cleaning the `Positions` column

```python
fifa_df.columns
```

```python
fifa_df["Positions"].unique()
```

```python
#Finding the number of positions
positions = set()
all_positions =fifa_df["Positions"].unique()
for pos in all_positions:
  position = re.findall(pattern=r"\w+", string=pos)
  positions.update(position)

print(f"There are {len(positions)} unique positions")
print(positions)
```

```python
# reordering the positions alphabetically
def reorderPositions(positions: str):
    splitted = re.findall(pattern=r"\w+", string=positions)
    splitted.sort()
    res = " ".join(splitted)
    return res

```

```python
fifa_df["Sorted Positions"]= fifa_df["Positions"].apply(reorderPositions)
```

```python
fifa_df["Sorted Positions"]
```

```python
fifa_df.drop("Positions", axis=1,inplace=True)
```
